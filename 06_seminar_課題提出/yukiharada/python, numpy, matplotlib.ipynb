<h1>python, numpy, matplotlib</h1>

```python
import logging
import sys

import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

logging.basicConfig(level=logging.INFO, stream=sys.stdout)
logger = logging.getLogger(__name__)
```

<h2>課題1</h2>

```python
first_word_chars = ['m', 'c', 'i', 'e', 'p', 'r', 'e', 't', 'o', ' ', 'o', 'o', 'i', 's', 'g', 'o', 'p']
second_word_chars = ['a', 'h', 'n', ' ', 'e', 'c', 'p', 'i', 'n', 'r', 'b', 't', 'c', '-', 'r', 'u', '!']

char_pairs = [char1 + char2 for char1, char2 in zip(first_word_chars, second_word_chars)]
text = ''
for char_pair in char_pairs:
    text += char_pair
    logger.debug({"char_pair": char_pair})
logger.info({"text": text})
```

<h2>課題2</h2>

```python
threshold  = 70

scores = np.array([40, 15, 72, 22, 43, 82, 75, 7, 34, 49, 95, 75, 85, 47, 63])

passed_students = sum(scores >= threshold)
z_scores = (scores - scores.mean()) / scores.std()
z_scores_mean = z_scores.mean()
z_scores_std = z_scores.std()

logger.debug({"z_scores_mean": z_scores_mean, "z_scores_std": z_scores_std})
logger.info({'passed_students': passed_students, 'z_scores': z_scores})
```

<h2>課題3</h2>

```python
def sigmoid(x):
    y = 1 / (1 + np.exp(-x))
    logger.debug({"action": "sigmoid", "x": x, "y": y})
    return y

def softmax(x):
    exp_x = np.exp(x)
    y = exp_x / np.sum(exp_x, axis=0)
    logger.debug({"action": "softmax", "x": x, "y": y})
    return y

def tanh(x):
    y = np.tanh(x)
    logger.debug({"action": "tanh", "x": x, "y": y})
    return y

x = np.arange(-5.0, 5.0, 0.1)

y_sigmoid = sigmoid(x)
y_softmax = softmax(x)
y_tanh = tanh(x)

fig, ax = plt.subplots()
ax.plot(x, y_sigmoid, label="Sigmoid")
ax.plot(x, y_softmax, label="Softmax")
ax.plot(x, y_tanh, label="Tanh")
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Sigmoid, Softmax, and Tanh Functions')
ax.grid(True)
plt.ylim(-1.2, 1.2)
plt.show()
```
